### Extracting Surfaces from subject MRIS

```python
#### IMPORTING MODULES ####
import os
import mne
mne.viz.set_3d_backend("notebook")
from osl_dynamics.meeg import rhino

#--------------------------------------------------------------------------------------#

#### DEFINING SURFACE EXTRACTION FUNCTIONION ####

def surface_extraction(mrifile, outdirsurf):

    # rhino.extract_surfaces(
    #    mri_file = mrifile,
    #    outdir = outdirsurf,
    # )

    rhino.extract_surfaces(
    mri_file=mrifile,
    outdir=outdirsurf,
    include_nose=True, #False for SQUID tutorial, True for OPM tutorial
    do_mri2mniaxes_xform=True, #False for SQUID tutorial, True for OPM tutorial
    show=True,
    )

    return None
```

```python
subjects = ["001"]

base = "/rdrives/DRS-foundation-brain/zoe_data/BIDS"
deriv = f"{base}/derivatives_anna_31032026"
surf_dir = f"{deriv}/anat_surfaces"
os.makedirs(deriv, exist_ok=True)

for subject in subjects:

    mri_file = (f"{base}/derivatives/FS_SUBJECTS/sub-{subject}/mri/T1.nii.gz")
    outdir_surf = (f"{surf_dir}/sub-{subject}")
    os.makedirs(outdir_surf, exist_ok=True)

    surface_extraction(mri_file, outdir_surf)
```
